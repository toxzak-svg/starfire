## 3. Phase 4 Experiment 1: GPU Baseline Benchmark

**Goal:** Establish single-GPU control condition

**Metrics:**
- Throughput (tokens/sec)
- Memory usage
- Forward/backward timing

**Expected runtime:** 20-30 minutes

```python
# Run GPU baseline benchmark
!python experiments/train_gpu_baseline.py \
    --device cuda \
    --models dense,hgsel \
    --batch-size 16 \
    --seq-length 128 \
    --d-model 256 \
    --d-ff 1024 \
    --num-layers 4 \
    --num-experts 64 \
    --warmup-steps 10 \
    --bench-steps 50 \
    --output results/gpu_baseline/colab_baseline.json
```

```python
# View results
import json
with open('results/gpu_baseline/colab_baseline.json', 'r') as f:
    results = json.load(f)

print("=" * 60)
print("GPU BASELINE RESULTS")
print("=" * 60)

parsed_model_results = {} # To store processed results for speedup calculation

# Process metadata if present
if 'metadata' in results:
    print("\nMETADATA:")
    for key, value in results['metadata'].items():
        print(f"  {key.replace('_', ' ').title()}: {value}")

print("\nMODEL BENCHMARKS:")
for model_metrics in results.get('results', []): # Iterate over the list under 'results' key
    model_name_kind = model_metrics.get('model_kind', 'unknown')
    throughput = model_metrics.get('throughput', {}).get('tokens_per_sec', 0)
    peak_memory = model_metrics.get('memory_mb', {}).get('peak_memory_mb', 0)
    avg_step_time = model_metrics.get('timings_ms', {}).get('avg_step_ms', 0)

    parsed_model_results[model_name_kind] = {
        'tokens_per_sec': throughput,
        'peak_memory_mb': peak_memory,
        'avg_step_ms': avg_step_time
    }

    print(f"\n{model_name_kind.upper()}:")
    print(f"  Throughput: {throughput:.1f} tokens/sec")
    print(f"  Peak Memory: {peak_memory:.1f} MB")
    print(f"  Avg Step Time: {avg_step_time:.2f} ms")

# Calculate speedup
if 'dense' in parsed_model_results and 'hgsel' in parsed_model_results:
    dense_throughput = parsed_model_results['dense']['tokens_per_sec']
    hgsel_throughput = parsed_model_results['hgsel']['tokens_per_sec']
    if dense_throughput > 0: # Avoid division by zero
        speedup = hgsel_throughput / dense_throughput
        print(f"\nHGSEL Speedup over DENSE: {speedup:.2f}x")
    else:
        print("\nCannot calculate speedup: Dense model throughput is zero.")
else:
    print("\nCannot calculate speedup: Both 'dense' and 'hgsel' results not found.")
```

## 6. Phase 4 Experiment 4: Memory Profiling

**Goal:** Understand memory breakdown

**Expected runtime:** 10-15 minutes

```python
# Run memory profiling experiment
import sys
sys.path.insert(0, '.')

from experiments.baselines.dense_transformer import TransformerModel
from hgsel.layer import HGSELLayer
from hgsel.distributed.memory_profiler import MemoryProfiler, estimate_model_memory_requirements

# Create models
print("Creating HGSEL model...")
model = TransformerModel(
    vocab_size=256,
    d_model=256,
    d_ff=1024,
    n_layers=4,
    n_heads=4,
    mlp_class=HGSELLayer
).cuda()

# Estimate memory
estimate = estimate_model_memory_requirements(
    model=model
)

print("\n" + "=" * 60)
print("MEMORY ESTIMATE")
print("=" * 60)
for key, value in estimate.items():
    print(f"{key}: {value}")
```
